### Getting rid of unmwanted / unuseful columns

```python
def remove_specific_airbnb_columns(df):
 
    individual_indices = [34, 36, 39, 40, 42, 43]
    
    range_1 = list(range(28, 34))  # Covers 28, 29, 30, 31, 32, 33
    range_2 = list(range(48, 62))  # Covers 48 through 61
    
    indices_to_drop = individual_indices + range_1 + range_2
    
    valid_indices = [i for i in indices_to_drop if i < len(df.columns)]
    
    cols_to_drop = df.columns[valid_indices]
    
    return df.drop(columns=cols_to_drop)

```

```python
df_Paris_Listings = remove_specific_airbnb_columns(df_Paris_Listings)
df_Lyon_Listings = remove_specific_airbnb_columns(df_Lyon_Listings)
df_Nice_Listings = remove_specific_airbnb_columns(df_Nice_Listings)
df_Bordeaux_Listings = remove_specific_airbnb_columns(df_Bordeaux_Listings)
df_Cannes_Listings = remove_specific_airbnb_columns(df_Cannes_Listings)
df_Antibes_Listings = remove_specific_airbnb_columns(df_Antibes_Listings)
df_Marseille_Listings = remove_specific_airbnb_columns(df_Marseille_Listings)
```

### Defining the listings columns of the cities


ttm_revenue_native - Total gross earnings in Euro (€) over the last 12 months.

ttm_avg_rate_native - Average Daily Rate (ADR) in Euro (€) charged per booked night.

ttm_occupancy - Percentage of available nights that were actually booked (Reserved÷Available).

ttm_revpar_native - Revenue Per Available Room in Euro (€); measures overall financial health.

ttm_reserved_days - Total number of nights guests stayed in the property.

ttm_blocked_days - Nights the host made unavailable (personal use, repairs, or maintenance).

ttm_available_days - Nights the property was listed and open for guests to book.

ttm_total_days - Sum of all days in the period (usually 365 days).

### Getting rid of unwanted / unusefull columns in historical datasets

```python
def clean_airbnb_historical(dff):
    indices = [5,6,7]
    cols_to_drop = [dff.columns[i] for i in indices if i < len(dff.columns)]
    return dff.drop(columns=cols_to_drop)
```

```python
df_Paris_Hisorical = clean_airbnb_historical(df_Paris_Hisorical)
df_Lyon_Historical = clean_airbnb_historical(df_Lyon_Historical)
df_Nice_Historical = clean_airbnb_historical(df_Nice_Historical)
df_Bordeaux_Historical = clean_airbnb_historical(df_Bordeaux_Historical)
df_Cannes_Historical = clean_airbnb_historical(df_Cannes_Historical)
df_Antibes_Historical = clean_airbnb_historical(df_Antibes_Historical)
df_Marseille_Historical = clean_airbnb_historical(df_Marseille_Historical)
```

### Drawing some basic insights

```python
# Most of the listings are entire homes/apartments, followed by private rooms, then hotel rooms, and finally shared rooms.

df_France_Listings.groupby('room_type')['City'].value_counts().sort_values()
```

```python
amenity_series = df_France_Listings['amenities'].dropna()

# Split on comma, strip spaces, explode into rows
amenity_series = (
    amenity_series
      .str.split(',')
      .explode()
      .str.strip()
)

# Get counts per amenity
amenity_counts = amenity_series.value_counts()

# show top 50 amenities
print(amenity_counts.head(50))
```

```python
# Top 100 highest revenue listings (adjust n as needed)
# support both possible column names to avoid KeyError
for candidate in ['ttm_revenue', 'ttm_revenue_native']:
    if candidate in df_France_Listings.columns:
        revenue_col = candidate
        break
else:
    raise KeyError(f"None of expected revenue columns found. Available columns: {df_France_Listings.columns.tolist()}")

top_revenue_listings = df_France_Listings.nlargest(100, revenue_col)

# Count amenities in top revenue group
top_amenity_counts = (
    top_revenue_listings['amenities']
    .dropna()
    .str.split(',')
    .explode()
    .str.strip()
    .value_counts()
)

print(top_amenity_counts.head(30))
```
